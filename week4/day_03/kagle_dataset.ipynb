```python
# Detect outliers
THRESHOLD = 3


def detect_outliers(data):
  outliers = []

  # get mean
  mean = np.mean(data)

  # get standard deviation
  std = np.std(data)

  # Apply Z score for each and every point in the datasets.
  for i in data:
    z_score = (i - mean) / std

    # check for outliers
    # append to outlier list in the case of outlier
    if np.abs(z_score) > THRESHOLD:
      outliers.append(i)
  return outliers
```

```python
outliers = detect_outliers(x['Purchase Amount (USD)'])
outliers
```

```python
df= x.sort_values(by='Purchase Amount (USD)')
# df.head()
print(df)
```

```python
sns.boxplot(df['Purchase Amount (USD)'])
```

```python
Q1, Q3 = np.percentile(df['Purchase Amount (USD)'], [25, 75])

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
```

```python
iqr = Q3 - Q1

print(f"IQR: {iqr}")
```

```python
lower_fence = Q1 - (1.5 * iqr)
upper_fence = Q3 + (1.5 * iqr)

print(f"Lower Fence: {lower_fence}")
print(f"Upper Fence: {upper_fence}")
```

```python
def detect_outlier_using_iqr(data, iqr, lower_fence, upper_fence):
  outliers = []
  for i in data:
      if isinstance(i, str):
          continue
      if i < lower_fence or i > upper_fence:
         outliers.append(i)

  return outliers
```

```python
outliers = detect_outlier_using_iqr(df['Purchase Amount (USD)'], iqr, lower_fence, upper_fence)
print(f'Detected Outliers:\n {outliers}')
```
